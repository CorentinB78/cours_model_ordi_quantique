```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import image
```

```python
feynman = image.imread("Feynman_quote.jpg")
feynman = np.mean(feynman, axis=2)  # grey scales
```

```python
print(feynman.shape)
plt.imshow(feynman, cmap='Grays_r')
plt.show()
```

```python
u, s, vh = np.linalg.svd(feynman)
```

```python
plt.plot(s[:])
plt.semilogy()
plt.show()
```

```python

lvl = 50
feynman_comp = (u[:, :lvl] * s[None, :lvl]) @ vh[:lvl, :]
# feynman_comp = (u[:, lvl][:, None] * s[lvl]) @ vh[lvl, :][None, :]

plt.imshow(feynman_comp, cmap='Grays_r')
plt.show()

plt.imshow(feynman, cmap='Grays_r')
plt.show()
```

```python
for N in [10, 20, 30, 40, 50]:
    for _ in range(10):
        mat = np.random.uniform(-1., 1., size=(N, N))
        s = np.linalg.svd(mat, compute_uv=False)
        
        plt.plot(s)

# plt.semilogy()
plt.show()
```
